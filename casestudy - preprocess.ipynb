```python
df.drop('No of Times Visited',axis=1,inplace=True)
```

```python
def impute_median(feature):
    median = df[feature].median()
    df[feature] = df[feature].fillna(median)
```

```python
cols=['Flat Area (in Sqft)','Sale Price','Lot Area (in Sqft)','No of Bathrooms','Area of the House from Basement (in Sqft)','Living Area after Renovation (in Sqft)','Longitude','Latitude','Zipcode']
```

```python
for col in cols:
    impute_median(col)
```

```python
df.isna().sum()
```

```python
for col in df:
    print(f"{col} has {df[col].nunique()}")
```

```python
#Outlier Detection
for i in ['Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'Area of the House from Basement (in Sqft)','Basement Area (in Sqft)',
       'Latitude', 'Longitude','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)']:
    Q1=df[i].quantile(0.25)
    Q2=df[i].quantile(0.50)
    Q3=df[i].quantile(0.75)
    iqr=Q3-Q1
    lower=Q1-(1.5*iqr)
    upper=Q3+(1.5*iqr)
    outliers_count = ((df[i] < lower) | (df[i] > upper)).sum()
    print(f'{i} : {outliers_count}')
```

```python
#handling outliers
for column in ['Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'Area of the House from Basement (in Sqft)','Basement Area (in Sqft)',
       'Latitude', 'Longitude','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)']:
       median = df[column].median()
       df[column] = np.where((df[column] < lower) | (df[column] > upper), median, df[column])
```

```python
df.drop('Date House was Sold',axis=1,inplace=True)
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
```

```python
cat_cols
```

```python
for i in cat_cols:
    print(f'{i} : {df[i].unique()}')
```

```python
# encoding "Waterfront View" Column
df.replace({'Waterfront View':{'No':0,'Yes':1}},inplace=True)

# encoding "Condition of the House" Column
df.replace({'Condition of the House':{'Excellent':1,'Good':2,'Okay':3,'Fair':4,'Bad':5}},inplace=True)
```

```python
x=df.drop('Sale Price',axis=1)
y=df['Sale Price']
```

```python
x1=x.copy()
x1.columns
```

```python
x1.drop(['Waterfront View','Condition of the House'],axis=1,inplace=True)
x1.head()
```

```python
#Scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x1=scaler.fit_transform(x1)
```

```python
x1=pd.DataFrame(x1,columns=['ID', 'No of Bedrooms', 'No of Bathrooms', 'Flat Area (in Sqft)',
       'Lot Area (in Sqft)', 'No of Floors', 'Overall Grade',
       'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
       'Age of House (in Years)', 'Renovated Year', 'Zipcode', 'Latitude',
       'Longitude', 'Living Area after Renovation (in Sqft)',
       'Lot Area after Renovation (in Sqft)'])
```

```python
x1.head()
```

```python
x2=x.drop(['ID', 'No of Bedrooms', 'No of Bathrooms', 'Flat Area (in Sqft)',
       'Lot Area (in Sqft)', 'No of Floors','Overall Grade',
       'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
       'Age of House (in Years)', 'Renovated Year', 'Zipcode', 'Latitude',
       'Longitude', 'Living Area after Renovation (in Sqft)',
       'Lot Area after Renovation (in Sqft)'],axis=1)
x=pd.concat([x1,x2],axis=1)
x.head()
```

```python
#model Evaluation
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42)
}
```

```python
r2_scores = {}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    r2_scores[name] = r2
```

```python
for name, score in r2_scores.items():
    print(f'{name} R-squared: {score:.4f}')
```
