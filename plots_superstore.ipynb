```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

BAR PLOT

```python

dff=df.groupby('Category')['Profit'].sum().reset_index()
plt.bar(dff['Category'],dff['Profit'],color='orange')
plt.title('Profit by category',fontsize=20)
plt.xlabel('Category',fontsize=16)
plt.ylabel('Profit',fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

PIE CHART

```python
df1=df.groupby('Region')['Sales'].sum()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df1,labels=df1.index,autopct='%1.2f%%',textprops={'fontsize': 14})
plt.title('Proportion of Sales by Region',fontsize=20)
plt.show()
```

HISTOGRAM

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Discount'], color='skyblue', edgecolor='black')
plt.title('Distribution of discount', fontsize=16)
plt.xlabel('Discount', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

HEATMAP

```python
df2=df[['Sales','Quantity','Discount','Profit']]
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df2.corr(),annot=True)
plt.show()
```

SCATTER PLOT

```python
plt.figure(figsize=(10,8))
plt.scatter(df2['Sales'],df['Profit'],c='r',edgecolor='k',s=100)
plt.title('sales v/s profit',fontsize=20)
plt.xlabel('Sales',fontsize=16)
plt.ylabel('Profit',fontsize=16)
plt.show()
```
